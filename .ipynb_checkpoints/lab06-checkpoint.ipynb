# Part 1

```python
# TODO define gaussian function here
def gaussian(x, mean, var):
    return (1/np.sqrt(2*np.pi*var))*np.exp(-(x-mean)**2/(2*var))
```

```python
# TODO plot gaussians here
gauss_approx = []
for i in range(0,10000):
    gauss_approx.append(np.random.normal())
plt.hist(gauss_approx, bins=np.arange(-5,5,0.25), density=True)
plt.show()
```

```python
# TODO fit gaussians, with known parameters, here
from scipy import optimize
x = np.arange(-5,5,0.001)
params, cov = optimize.curve_fit(gaussian, x, gauss_approx)
print(params)
# TODO plot result
plt.hist(gauss_approx, bins=np.arange(-5,5,0.25), density=True)
plt.plot(x,gaussian(x, params[0], params[1]))
```

```python
# TODO: load data here
HDdata = np.loadtxt('HDalpha_data')
print(HDdata.shape)
```

```python
# TODO: fit a function to this data and plot it
def double_gaussian(x, mean1, var1, amp1, mean2, var2, amp2):
    return amp1*(1/np.sqrt(2*np.pi*var1))*np.exp(-(x-mean1)**2/(2*var1)) + amp2*(1/np.sqrt(2*np.pi*var2))*np.exp(-(x-mean2)**2/(2*var2))
params, cov = optimize.curve_fit(double_gaussian, HDdata[:,0], HDdata[:,1])
plt.plot(HDdata[:,0],HDdata[:,1])
plt.plot(HDdata[:,0], double_gaussian(HDdata[:,0], *params))
plt.show()
```

# Part 3

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
x_pos = np.array([10,-10])
y_pos = np.array([15,6])
Q_vals = np.array([3,-3])
x_array = np.linspace(-31.5, 31.5, 64)
y_array = np.linspace(-31.5, 31.5, 64)
# TODO get flux
flux = flux_for_dipole(x_pos, y_pos, Q_vals, x_array, y_array)
# TODO plot contours of the flux
plt.contour(x_array,y_array,flux,40)
```
